## Tiền xử lý dữ liệu

```python
import pandas as pd
```

```python

# Đọc file dữ liệu 
df = pd.read_excel('Thay-Thong-Diem-Tong-Hop-K55_fix_3.xls', sheet_name = "TongHopHocCayQD", dtype={'DA':'|S', 'Class-XLTN':'|S'})
df.head(5)
df = df[:314]
df
```

```python
(df['A4'].describe()) # quan sát mô tả về 1 trưởng dữ liệu
```

```python
# df['Class-XLTN']
```

```python
# df.columns.values.tolist()
```

```python
# Chỗ này lấy các trường dữ liệu bằng list các tên - có thể dùng iloc để lấy theo index column
new_df = df[['STT',
 'MaSV',
 'D2',
 'C2',
 'B2',
 'A2',
 'D3',
 'C3',
 'B3',
 'A3',
 'D4',
 'C4',
 'B4',
 'A4',
 'DA',
 'Class-XLTN']]
```

```python
# Dùng cho việc mã hoá lại dữ liệu
from sklearn.preprocessing import LabelEncoder
le_DA = LabelEncoder()
le_Class_XLTN = LabelEncoder()
```

```python
# Mã hoá cho trường DA, XLTN -> đồ án và xếp loại tốt nghiệp
new_df['DA_n'] = le_DA.fit_transform(new_df['DA'])
new_df['XLTN_n'] = le_Class_XLTN.fit_transform(new_df['Class-XLTN'])
# a = new_df['DA'].astype(str)
# a
new_df
```

```python
conver_df = new_df.drop(['DA', 'Class-XLTN','MaSV'], axis =1)
# conver_df = conver_df.drop(315, axis=0)
```

```python
# Dữ liệu đã được mã hoá và có thể train mô hình
conver_df
```

```python
# Các quan sát và nhãn là inputs_l và target tương ứng
inputs_l = conver_df.drop(['STT','XLTN_n'], axis = 1)
target = conver_df['XLTN_n']
# inputs_l
```

```python
# target
```

## Train model

```python
# Chia dữ liệu tành tập train và tập test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_l, target, test_size =0.33)
```

```python
# import tree từ lib sklearn và huấn luyện mô hình với tham số entropy tương ứng thuât toán id3
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion='entropy')

model.fit(X_train, y_train) # train trên tập huấn luyện
```

```python
inputs_l
```

```python
# test mô hình trên tập test - Đánh giá mô hình
model.score(X_test, y_test)
```
