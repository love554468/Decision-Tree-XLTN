```python
new_df = df[['STT',
 'MaSV',
 'D2',
 'C2',
 'B2',
 'A2',
 'D3',
 'C3',
 'B3',
 'A3',
 'D4',
 'C4',
 'B4',
 'A4',
 'DA',
 'Class-XLTN']]
```

```python
from sklearn.preprocessing import LabelEncoder
le_DA = LabelEncoder()
le_Class_XLTN = LabelEncoder()
```

```python
new_df['DA_n'] = le_DA.fit_transform(new_df['DA'])
new_df['XLTN_n'] = le_Class_XLTN.fit_transform(new_df['Class-XLTN'])
# a = new_df['DA'].astype(str)
# a
```

```python
conver_df = new_df.drop(['DA', 'Class-XLTN','MaSV'], axis =1)
conver_df = conver_df.drop(315, axis=0)
```

```python
conver_df
```

```python
inputs_l = conver_df.drop(['XLTN_n'], axis = 1)
target = conver_df['XLTN_n']
inputs_l
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(inputs_l, target)
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
a =  new_df.columns.values.tolist()
f = a[:-4]
t = np.array(['C', 'B', 'A'])
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   feature_names=f,  
                   class_names=t,
                   filled=True)
```
