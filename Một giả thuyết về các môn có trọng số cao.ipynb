```python
 dic_gia_thiet= dict() # Xem xét giả thiết đặt ra có đúng không tính tỉ lệ
for i in range(50):
    mon_duoc_A = df[df.iloc[:,i] == 0]
    a = df[df.iloc[:,i] == 0].shape[0] # số lượng thì sinh mà có môn i được A
    mon_A_XLTN_A = mon_duoc_A[df['Xếp loại TN']==0]
    b = mon_A_XLTN_A.shape[0]                      
#     x = (mon_duoc_A,mon_A_XLTN_A)
    dic_gia_thiet[i] = round((b/a) *100,2)
```
